Princípio da Superposição
---

Segundo o Princípio da Superposição para circuitos elétricos, a tensão ou a corrente em um elemento em um circuito linear é dada pela soma algébrica das tensões ou correntes devido a atuação isolada de cada uma das fontes independentes que compõem o circuito. (2)
Nota-se que os circuitos divisores de tensão, Figura 5 e Figura 6, são constituídos unicamente por elementos resistivos, que são lineares. Portanto, nesses circuitos é possível aplicar o Princípio da Superposição. Já o circuito retificador de meia, Figura 8, há a presença de um diodo, que é um componente eletrônico não linear, como observamos na Figura 10, existem duas regiões de operação distintas nas quais o componente atua.



image.png

A característica I/V do diodo apresenta comportamento não linear e, assim, não satisfaz o Princípio da Superposição, consequentemente, o circuito retificador de meia onda não está sujeito a esse princípio e, consequentemente, o circuito de filtragem também não estará, visto que ele está conectado à saída do diodo. Feitas tais considerações a respeito dos elementos que compõe o circuito, é possível aproximar a tensão sobre o resistor $R_3$ no ponto Sinal 2 como sendo a soma dos efeitos de cada fonte independente sobre o elemento resistivo em questão. Observa-se que o circuito da Figura 7 é formado unicamente por resistores, que seguem a Lei de Ohm e não induzem mudanças de fase nas tensões senoidais

$V = RI \tag{1}$

Por esse motivo, a tensão senoidal no ponto Sinal 2 referentes às fontes independentes podem ser somadas, Princípio da Superposição aplicado ao elemento $R_3$. Nesse caso, apenas a amplitude da senoide será diminuída proporcionalmente ao valor das resistências, segundo a Lei de Ohm. O Gráfico 3 apresenta de forma esquemática a forma de onda referente à tensão esperada no ponto Sinal 2.

Diodo 
---
O diodo, como dito, apresenta uma característica não linear e, agora, iremos melhor analisar essa característica. Em componentes lineares, como resistores, a corrente é diretamente proporcional à variação da tensão, podendo ser aproximada por uma equação do tipo

$I = aV \tag{2}$

onde $a$ é a taxa de variação da reta. Observa-se que a Lei de Ohm dada $(1)$ é uma função do tipo $(2)$.

Em dispositivos não lineares, por sua vez, a corrente não é diretamente proporcional a tensão. Em particular, os diodos apresentam uma relação tipicamente quadrática que pode ser aproximada da forma

$I = aV + bV^2 \tag{3}$

onde o termo linear corresponde a uma componente DC (DC bias). (4)


image.png

Como dito, a tensão aplicada ao anodo do diodo é dada pelo Princípio da Superposição e, aqui, aproximada para 

$V_{S2} = V_P + V_M \tag{4}$

Subsituindo a expressão $4$ em $3$, temos:

$I = a(V_P + V_M) + b(V_P + V_M)^2$

$I = a(V_P + V_M) + b(V_P^2 + 2V_PV_M + V_M^2) \tag{5}$

Como $V_P$ e $V_M$ são tensões senoidais, temos:

$V_P = v_Psen(\omega_Pt), \omega_P = \frac{2\pi}{f_P}$

e

$V_M = v_Msen(\omega_Mt), \omega_M = \frac{2\pi}{f_M}$

Substituindo $V_P$ e $V_M$ em $(5)$:

$I = av_Psen(\omega_{P}t) av_Msen(\omega_{M}t) + bv_P^2sen^2(\omega_{P}t)+2v_Pv_Msen(\omega_{P}t)sen(\omega_{M}t)+bv_M^2sen^2(\omega_{M}t) \tag{6}$

Usaremos a identidade trigonométrica $sen^2(x) = \frac{1-cos(2x)}{2}$ e substituiremos em $(6)$:

$I = av_Psen(\omega_Pt)+av_Msen(\omega_Mt)+\frac{bv_P^2\left [1-cos(2\omega_Pt)  \right ]}{2}+2bv_Pv_Msen(\omega_Pt)sen(\omega_Mt)+\frac{bv_M^2\left [1-cos(2\omega_Mt)  \right ]}{2} \tag{7}$


Analisando a expressão $(7)$, nota-se que o primeiro termo corresponde ao sinal $V_P$, que chamaremos de portadora, o segundo termo corresponde ao sinal $V_M$, que chamaremos de modulante. O quarto termo de $(7)$ apresenta o produto de $V_P$ e $V_M$, então, a tensão o resistor $R_5$ não ser será simplesmente decorrente do Princípio da Superposição antes aplicado ao $R_3$. O Gráfico 4 ilustra a tensão sobre o resistor $R_5$ na situação hipotética na qual se considera a corrente através do diodo ignorando o seu efeito de retificação, faz-se isso para enfatizar que no ponto Sinal 3 a tensão não é a mesma do ponto Sinal 2. (4)

Máxima Transferência de Potência
---

Em um circuito linear, como o da Figura 7, a máxima transferência de potência que o circuito pode fornecer a carga $R_3$ pode ser determinada com o auxílio de um circuito equivalente Thévenin. Para calcular a resistência de Thévenin, $R_{TH}$ nos terminais de $R_3$, as fontes de tensão são substituídas por suas resistências internas, consideradas nulas no modelo de fonte ideal adotado. Feitas essas considerações, nota-se que $R_{TH}$ é, na verdade, a associação dos resistores $R_1$ e $R_2$ paralelos.



image.png

image.png

A partir do rearranjo ilustrado na Figura 13, nota-se que a resistência de Thévenin é dada pela expressão

$R_{TH} = \frac{R_1R_2}{R1 + R_2}$

Portanto, o circuito equivalente Thévenin é esquematicamente representado na Figura 14.

image.png

Pode-se, agora, calcular a potência dissipada por $R_3$ aplicando a o divisor de tensão sobre ele e a Lei de Kirchhoff na malha.
 
$P_{R_3} = I^{2} R_{TH} = \left ( \frac{V_{TH}}{R_{TH} + R_3} \right )^2R_3$
 
Fixando os valores das fontes e dos resistores, calcula-se a derivada da potência em relação a $R_3$, o ponto de máxima transferência de potência corresponde ao ponto onde a derivada se anula. (2)
 
$\frac{dP_{R_{3}}}{dR_{3}} = (V_{TH})^{2}\left [ \frac{(R_{TH} + R_3) - 2R_3\left ( R_{TH}+R_3 \right )}{\left ( R_{TH} + R_3 \right )^4} \right ] = V_{TH}^{2}\left [ \frac{\left ( R_{TH} - R_3 \right )}{\left ( R_{TH}+R_3 \right )^{3}} \right ] = 0$
 
$\Rightarrow R_{TH} - R_{3} = 0$
 
 
$\Rightarrow R_{TH} = R_{3}$
 
E, assim, para que haja a máxima transferência de potência no resistor $R_3$ temos que:
 
$R_{TH} = R_3 = \frac{R_{1}R_{2}}{R_1 + R_2}$
 
Adicionalmente, nota-se também, por meio das Figura 5 e Figura 6, que o valor de $R_3$ influencia a amplitude da tensão no ponto Sinal 2. Por se tratar de um componente ôhmico, maiores valores de resistência implicam em maior queda de tensão sobre $R_3$, o que aumenta a tensão no ponto Sinal 2; menores valores de $R_3$ diminuem a queda de tensão sobre o componente, o que faz diminuir a amplitude da tensão no ponto Sinal 2. Outra forma de interpretar o mesmo efeito é por meio da Figura 7, para menores valores de $R_3$, maior a corrente drenada para o terra do circuito e, consequentemente, menor a tensão no ponto Sinal 2; para maiores valores de $R_3$, menor a corrente drenada para o terra através desse resistor e maior a tensão no Sinal 2.

A Figura 8 apresenta o sub circuito regulador que contém o diodo $D_1$. No intervalo de polarização direta, há fluxo de corrente no circuito $LC$ paralelo, o que faz com que o indutor e o capacitor armazenem energia, porém existe também uma troca mútua de energia entre esses componentes, o que dá origem a uma forma de onda como a apresentada no Gráfico 5, mas negativa, nos mesmos instantes de tempo. Assim, o sinal de saída é como o ilustrado no Gráfico 6. (4)

Ademais, os elementos reativos, indutor, $L_1$, e capacitor, $C_1$ têm suas reatâncias alteradas em função da frequência, o caso em que suas reatâncias apresentam o mesmo valor absoluto é chamada de ressonância. Os circuitos ressonantes possuem aplicações que exigem certa seletividade em termos de frequência, já que as reatâncias variam com ela. Além disso, na ressonância, o circuito $RLC$ passa a apresentar impedância puramente resistiva. (2)
O cálculo da reatância do circuito $LC$ é apresentado abaixo:

$Z = jZ_{L}//\frac{-j}{Z_{C}} = \frac{jZ_L\left ( -jZ_{C} \right )}{jZ_{L}-jZ_{C}} = \frac{Z_{L}Z_{C}}{j\left ( Z_{L}-Z_{C} \right)} = -j\frac{Z_{L}Z_{C}}{Z_{L}-Z_{C}}$


$\therefore  Z = -j\frac{Z_{L}Z_{C}}{Z_{L}-Z_{C}} \tag{8}$

Nota-se que, na condição de ressonância, $Z_{L} = Z_{C}$  e isso faz com que a impedância $Z \rightarrow \infty$, ou seja, é como se os componentes $L_1$ e $C_1$ fossem removidos do circuito, permanecendo apenas o $R_4$.
Observando a $(8)$ que define $Z$, vemos que, para frequências abaixo da frequência de ressonância, a impedância capacitiva é maior do que a indutiva, já que a $Z_{C}$ é dado pelo inverso da frequência, consequentemente, para valores da frequência acima da frequência de ressonância, $Z$ apresenta característica predominantemente indutiva.
Ademais, examinando a condição de ressonância, nota-se que a frequência de ressonância é dada pela expressão

$\omega_LL = \frac{1}{\omega_CC}$

como $\omega = 2\pi f$, temos:

$f^2 = \frac{1}{(2\pi)^2LC}$

$f = \frac{1}{2\pi\sqrt{LC}}$

Resposta em Frequência
---

Para determinar a resposta em frequência do circuito, faz-se uso da função de transferência $\mathbf{H}(\omega)$, sendo a resposta em frequência o gráfico de $\mathbf{H}(\omega)$ por $\omega$ no intervalo $0 \leqslant \omega < \infty$. A função de transferência será definida como a razão entre uma saída fasorial $\mathbf{Y}(\omega)$ e uma entrada fasorial $\mathbf{X}(\omega)$, isto é:

$\mathbf{H}(\omega) = \frac{\mathbf{Y}(\omega)}{\mathbf{X}(\omega)} \tag{9}$

Para o circuito LC, Figura 8

$Z_{eq} = R + j \omega L + \frac{1}{j \omega C}$

$Z_{eq} = R + j \omega L - j\frac{1}{ \omega C}$

$Z_{eq} = R +  \frac{j \omega^{2} LC - j}{\omega C}$

$Z_{eq} = R +  j\left (\frac{ \omega^{2} LC - 1}{\omega C}  \right )$

$Z_{eq} = R -  j\left (\frac{ 1 - \omega^{2} LC}{\omega C}  \right )$

$Z_{eq} = R +  j\left (\frac{ \omega C}{1 - \omega^{2} LC}  \right ) \tag{10}$

Aplicando a Lei de Kirchhoff para malhas: 

$V_{out} = V_{S3} - V_R = V_{S3} - RI \tag{11}$

Como 

$I = \frac{V_{S3}}{Z_{eq}} \tag{12}$

Substituindo $(12)$ em $(11)$, temos:

$V_{out} = V_{S3} - \frac{R}{Z_{eq}}V_{S3} = V_{S3}\left ( 1 - \frac{R}{Z_{eq}} \right ) \tag{13}$

$(13)$ em $(15)$:

$\mathbf{H}(\omega) = \frac{V_{S3}\left ( 1 - \frac{R}{Z_{eq}} \right )}{V_{S3}} = 1 - \frac{R}{Z_{eq}} \tag{14}$

Substituindo $(11)$ em $(15)$:

$\mathbf{H}(\omega)= 1 - \frac{R}{j\frac{\omega C}{1 - \omega^{2}LC}}$

$\mathbf{H}(\omega) = 1 - R \frac{\left ( 1 - \omega^{2} LC \right )}{j \omega C}$

$\mathbf{H}(\omega) = 1 +j\frac{R\left ( 1-\omega^{2}LC \right )}{\omega C} \tag{15}$

Calculando o módulo de $(15)$:

$\left |\mathbf{H}(\omega)  \right | =\sqrt{ 1 + \left [ \frac{R\left ( 1 - \omega^{2}LC \right )}{\omega C} \right ]^{2}}$

O ânguo de fase $\phi$ é dado por:

$\phi = tg^{-1}\left ( \frac{\frac{R\left ( 1 - \omega^{2}LC \right )}{\omega C}}{1} \right ) = tg^{-1}\left ( \frac{R\left ( 1 - \omega^{2}LC \right )}{\omega C} \right )$

$ \therefore  \phi =tg^{-1} \left ( \frac{R\left ( 1 - \omega^{2}LC \right )}{\omega C} \right )$
